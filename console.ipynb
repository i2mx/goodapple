```python
import scipy as sp
from scipy.io import wavfile 
from scipy.fftpack import fft, fftshift, fftfreq
import matplotlib.pyplot as plt
from scipy import signal
import numpy as np
```

```python
notes = []
with open("notes.txt") as file:
    for line in file:
        notes.append(float(line.strip()))

notes
```

```python
samplerate, data = wavfile.read("【東方】Bad Apple!! ＰＶ【影絵】 [FtutLA63Cp8].wav")

data = data[70000:80000]

plt.plot(data)
```

```python
fft_out = fft(data)
plt.plot(np.abs(fft_out))
```

```python
f, t, Sxx = signal.spectrogram(data, 10e3)
plt.pcolormesh(t, f, Sxx, shading='gouraud')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
